```python
# response = requests.get(url)
```

```python
soup = bs(response.text, 'html.parser')
```

```python
data = soup.text # one big string of web text
```

```python
jieba.add_word('机器之心')
jieba.add_word('光大')
jieba.add_word('光大控股')
jieba.add_word('特斯联')
```

```python
seg_list = jieba.cut(data, cut_all=False)
print("Full Mode: " + "/ ".join(seg_list))
```

```python
words = pseg.cut(data)
l = [(word, flag) for word, flag in words]
```

```python
l
```

```python
for item in l:
    if item[1] == 'v':
        print(item)
```

```python
jieba.analyse.textrank(data, topK=200, withWeight=False, allowPOS=('ns', 'n', 'vn', 'v'))
```
